```python
df = pd.read_csv("F:\\AIX Docs\\Data science\\Assignments\\Telecom retention_Telco Company.csv")
```

```python
df1 = df*1
```

```python
X = df1[['Total day charge','Total eve charge','Total night charge','Total intl charge']]
```

```python
y = df1['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
X_train.shape
```

```python
y_train.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_train = LogisticRegression()
```

```python
model_train = model_train.fit(X_train,y_train)
```

```python
pred = model_train.predict(X_test)
pred
```

```python
### Confustion Matrix
## Actual Data ## predicted data
df2 = pd.DataFrame({'Actual_Data': y_test, 'Predicted_Data': pred})
df2
```

```python
from sklearn.metrics import confusion_matrix
```
